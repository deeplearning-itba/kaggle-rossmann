```python
import numpy as np
import keras

class DataGenerator(keras.utils.Sequence):
    def get_num_of_docs(self):
        df = pd.read_csv(self.file_path, iterator=True, chunksize=1000)
        for i, chunk in enumerate(df):
            pass
        return i * 1000 + len(chunk)
    
    def __init__(self, file_path, batch_size=1000, shuffle=True):
        self.file_path = file_path
        self.batch_size = batch_size
        self.shuffle = shuffle
        self.num_of_docs = self.get_num_of_docs()
        self.on_epoch_end()

    def __len__(self):
        'Denotes the number of batches per epoch'
        return int(np.floor(self.num_of_docs / self.batch_size))

    def __getitem__(self, index):
        'Generate one batch of data'
        # Generate indexes of the batch
        indexes = self.indexes[index*self.batch_size:(index+1)*self.batch_size]

        # Find list of IDs
        list_IDs_temp = [self.list_IDs[k] for k in indexes]

        # Generate data
        X, y = self.__data_generation(list_IDs_temp)

        return X, y

    def on_epoch_end(self):
        'Updates indexes after each epoch'
        self.indexes = np.arange(self.num_of_docs)
        if self.shuffle == True:
            np.random.shuffle(self.indexes)

    def __data_generation(self, list_IDs_temp):
        'Generates data containing batch_size samples' # X : (n_samples, *dim, n_channels)
        batch = []
        for i in list_IDs_temp:
            batch.append(list(pd.read_csv(PATH, skiprows=i, nrows=1)))
        
        batch = np.array(batch)

        return batch[0, :], batch[5, :]
```
